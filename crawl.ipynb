```python
import time
import requests
import pandas as pd
from datetime import datetime
```

```python
lat, lon = 21.0285, 105.8542


features = [
    "T2M", "ALLSKY_SFC_SW_DWN", "RH2M", "WS2M", "T2MDEW",
    "T2MWET", "PS", "WS50M", "ALLSKY_SFC_LW_DWN", "WD50M", "WD2M", "ALLSKY_SFC_SW_DIFF",
    "ALLSKY_SFC_SW_DNI", "ALLSKY_SRF_ALB", "CLOUD_AMT", "PRECTOTCORR", "TOA_SW_DWN", "WS10M",
    "GWETPROF", "TS", "WD10M", "ALLSKY_SFC_PAR_TOT", "QV2M", 
    "ALLSKY_KT", "CLRSKY_SFC_SW_DWN", "GWETROOT", "GWETTOP"
]

parameters1 = ",".join(features[:14])
parameters2 = ",".join(features[14:])
len(features)
```

**Nasa chỉ cho phép lấy dữ liệu hourly theo từng năm => chia nhỏ**

```python
df = []
for i in range(2001, 2025):
    time.sleep(10)
    # 14 features đầu
    url = f"https://power.larc.nasa.gov/api/temporal/hourly/point?start={i}0101&end={i}1231&latitude={lat}&longitude={lon}&community=SB&parameters={parameters1}"
    response = requests.get(url, timeout=120, stream=True)
    data = response.json()
    if (response.status_code == 200):
        print("success", 1, i)
    
    df1 = pd.DataFrame(data["properties"]["parameter"])
    df1.index = pd.to_datetime(df1.index, format="%Y%m%d%H")
    df1.index.name = "Date Time"
    df1.reset_index(inplace=True)
    

    # lấy các feature còn lại
    url = f"https://power.larc.nasa.gov/api/temporal/hourly/point?start={i}0101&end={i}1231&latitude={lat}&longitude={lon}&community=SB&parameters={parameters2}"
    response = requests.get(url, timeout=120, stream=True)
    data = response.json()
    if (response.status_code == 200):
        print("success", 2, i)
    
    df2 = pd.DataFrame(data["properties"]["parameter"])
    df2.index = pd.to_datetime(df2.index, format="%Y%m%d%H")
    df2.index.name = "Date Time"
    df2.reset_index(inplace=True)
    

    df_merge = pd.merge(df1, df2, on="Date Time", how="outer")

    df.append(df_merge)
```

```python
df = df[:13]


for i in range(2013, 2025):
    time.sleep(10)
    # 14 features đầu
    url = f"https://power.larc.nasa.gov/api/temporal/hourly/point?start={i}0101&end={i}1231&latitude={lat}&longitude={lon}&community=SB&parameters={parameters1}"
    response = requests.get(url, timeout=120, stream=True)
    data = response.json()
    if (response.status_code == 200):
        print("success", 1, i)
    else:
        print("error", 1, i)
        break
    df1 = pd.DataFrame(data["properties"]["parameter"])
    df1.index = pd.to_datetime(df1.index, format="%Y%m%d%H")
    df1.index.name = "Date Time"
    df1.reset_index(inplace=True)
    

    # lấy các feature còn lại
    url = f"https://power.larc.nasa.gov/api/temporal/hourly/point?start={i}0101&end={i}1231&latitude={lat}&longitude={lon}&community=SB&parameters={parameters2}"
    response = requests.get(url, timeout=120, stream=True)
    data = response.json()
    if (response.status_code == 200):
        print("success", 2, i)
    else:
        print("error", 2, i)
        break
    df2 = pd.DataFrame(data["properties"]["parameter"])
    df2.index = pd.to_datetime(df2.index, format="%Y%m%d%H")
    df2.index.name = "Date Time"
    df2.reset_index(inplace=True)
    

    df_merge = pd.merge(df1, df2, on="Date Time", how="outer")

    df.append(df_merge)
```
